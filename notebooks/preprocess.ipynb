```python
import pandas as pd
```

```python
def aug(text):
    sts = text.split('。')
    for i, st in enumerate(sts):
        if st == '':
            continue
        if st[-1] not in ['。', '！', '？']:
            sts[i] = st + '。'
        if st[0] in ['A', 'B', 'C', 'D']:
            sts[i] = sts[i][1:]
            if sts[i][0] in ['、', '．']:
                sts[i] = sts[i][1:]
    while '' in sts:
        sts.remove('')
    return sts


train_df = pd.read_csv('../data-org/train.csv', sep='\t', index_col='id')

# data = train_df[500:600]
# data['prediction'] = torch.argmax(output_tensors[0], 1)
# data = data[['label', 'prediction', 'text']]
# with pd.option_context('display.max_rows', None, 'display.max_columns', None):
#     display(data[data.label != data.prediction])


long_text_df = train_df[train_df.text.map(len) > 126]
long_text_ids = long_text_df.index
display(long_text_df)

split_texts = [aug(train_df[train_df.index == idx].text.values[0]) for idx in long_text_ids]

new_labels = [None] * len(long_text_ids)
new_labels[0] = [0, 1, 0, 1]
new_labels[1] = [0, 1, 0]
new_labels[2] = [1]
new_labels[3] = [0, 1, 1]
new_labels[4] = [1]
new_labels[5] = [1, 0, 0]
new_labels[6] = [1, 1, 1]
new_labels[7] = [1, 1, 0, 0]
new_labels[8] = [0, 1]

new_labels = [x for x in new_labels]

for j, idx in enumerate(long_text_ids):
    train_df.drop(index=idx, inplace=True)
    for k in range(len(new_labels[j])):
        train_df = pd.concat((train_df, pd.DataFrame([{'id':str(idx)+'-'+str(k),'label':new_labels[j][k], 'text':split_texts[j][k]}]).set_index('id')))

display(train_df[train_df.text.map(len) > 126])
train_df.to_csv('../data/train.csv', sep='\t')
```
